```python
import pandas as pd
```

```python
df = pd.read_csv("SuperMarket.csv")
```

```python
df.info()
```

```python
print("Missing value per column", df.isnull().sum())
```

```python
print("Duplicate each row is:", df.duplicated().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python

sns.histplot(df['Sales'], kde =True)
plt.title('Distribution of Total Sales')
plt.xlabel('Total Purchase Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(df['Profit'], kde = True)
plt.title('Distribution of Profit')
plt.xlabel('Total Profit')
plt.ylabel ('Frequency')
plt.show()
```

```python
sns.boxplot(x=df['Sales'])
plt.title('Distribution of Total Sales')
plt.xlabel('Total Purchase Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x=df['Profit'])
plt.title('Distribution of Total Profit')
plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Region'].value_counts().plot(kind='bar')
df.describe()
```

```python
category_summary = df.groupby('Category')['Sales'].sum().reset_index()

sns.barplot(x='Sales', y='Category', data=category_summary)
plt.title('Total Sales by Category')
plt.xlabel('Total Sales')
plt.ylabel('Category')
plt.show()
```

```python
category_summary = df.groupby('Sub Category')['Sales'].sum().reset_index()

sns.barplot(x='Sales', y='Sub Category', data=category_summary)
plt.title('Total Sales by Sub Category')
plt.xlabel('Total Sales')
plt.ylabel('Sub Category')
plt.show()
```

```python
df.groupby('Order Date')['Sales'].sum().plot(figsize=(12,5), title='Sales Over Time')
```
